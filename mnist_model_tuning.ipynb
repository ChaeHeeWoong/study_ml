# Data 불러오기

```python
import numpy as np
import pandas as pd
```

```python
mnist_train = pd.read_csv('/Users/chaehuiung/Desktop/project/work/mnist/data/train.csv')
mnist_test = pd.read_csv('/Users/chaehuiung/Desktop/project/work/mnist/data/test.csv')
```

# Model 후보군


## Logistic Regression, L-GBM, Random Forest
## DNN, CNN

```python
from sklearn.model_selection import train_test_split

train, valid = train_test_split(mnist_train, test_size=0.2, random_state=0)

y_train=train['label']
x_train=train.drop(['label'], axis=1)

x_test= mnist_test
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

lr = LogisticRegression(n_jobs=-1)
lr.fit(x_train, y_train)
```

```python
scores = cross_val_score(lr, x_train, y_train, cv=5,n_jobs=-1,verbose=2) # model, train, target, cross validation

print('cross-val-score \n{}'.format(scores))
print('cross-val-score.mean \n{:.3f}'.format(scores.mean()))
```

```python
#test
pred_y = lr.predict(x_test)
```

```python
from pandas import DataFrame

sol = DataFrame(pred_y, columns= ['Label'], index = [list(range(1,28001))])
```

```python
sol.to_csv('test_log_regression.csv')
```

# L-GBM

```python
from lightgbm import LGBMClassifier

lc = LGBMClassifier() #hyperparameter : learning_rate
lc.fit(x_train, y_train)
```

```python
scores = cross_val_score(lc, x_train, y_train, cv=5,n_jobs=-1,verbose=2) # model, train, target, cross validation

print('cross-val-score \n{}'.format(scores))
print('cross-val-score.mean \n{:.3f}'.format(scores.mean()))
```

```python
#test
pred_y = lc.predict(x_test)
```

```python
from pandas import DataFrame

sol = DataFrame(pred_y, columns= ['Label'], index = [list(range(1,28001))])
```

```python
sol.to_csv('test_light_gbm.csv')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier() #hyperparameter : learning_rate
rf.fit(x_train, y_train)
```

```python
scores = cross_val_score(rf, x_train, y_train, cv=5,n_jobs=-1,verbose=2) # model, train, target, cross validation

print('cross-val-score \n{}'.format(scores))
print('cross-val-score.mean \n{:.3f}'.format(scores.mean()))
```

```python
#test
pred_y = lc.predict(x_test)

from pandas import DataFrame

sol = DataFrame(pred_y, columns= ['Label'], index = [list(range(1,28001))])
sol.to_csv('test_randomforest.csv')
```

# DNN

```python
import tensorflow as tf
```

```python
feature_cols = tf.contrib.learn.infer_real_valued_columns_from_input(x_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units = [300,100], n_classes=10, feature_columns = feature_cols)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf)
dnn_clf.fit(x_train, y_train, batch_size=50, steps=40000)
```

```python
y_pred = dnn_clf.predict(x_test)
```

```python
len(y_pred['classes'])
```

```python
from pandas import DataFrame

sol = DataFrame(y_pred['classes'], columns= ['Label'], index = [list(range(1,28001))])
sol.to_csv('test_dnn.csv')
```
